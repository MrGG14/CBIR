# CBIR CON HISTOGRAMA DE TEXTURAS

```python
from collections import Counter
import matplotlib.pyplot as plt
import cv2
import random
import numpy as np
import os
from sklearn.neighbors import NearestNeighbors
from skimage.feature import graycomatrix, graycoprops  # Cambio de import

# Función para calcular las características de textura de Haralick
def calcular_histograma_textura(imagen):
    gray_imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
    
    # Calcular la matriz de co-ocurrencia de niveles de gris (GLCM) utilizando graycomatrix
    glcm = graycomatrix(gray_imagen, [1], [0], symmetric=True, normed=True)
    
    # Calcular las características de textura de Haralick utilizando graycoprops
    textura = graycoprops(glcm, prop='contrast')[0]
    
    return textura


def get_n_similar(n, counts, reverse):
    sorted_counts = sorted(counts.items(), key=lambda x:x[1], reverse=reverse)[:n] #Ordeno y me quedo con los 5 más parecidos
    sorted_indexes = [idx[0] for idx in sorted_counts]
    return sorted_indexes

def display_n_similar(sorted_indexes, n):
    if n <= 0:
        print("El valor de n debe ser mayor que 0.")
        return

    if n > len(sorted_indexes):
        print(f"Solo hay {len(sorted_indexes)} imágenes en la lista. Mostrando todas.")
        n = len(sorted_indexes)
    fig, axs = plt.subplots(1, n, figsize=(15,15))
    for i in range(n):
        n_carpeta = (sorted_indexes[i][0])
        n_img = (sorted_indexes[i][1])
        path = images_path[(n_carpeta, n_img)]
        im = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)
        axs[i].imshow(im)
        axs[i].axis('off')
        titulo = f'{n_carpeta}_{n_img}'
        axs[i].set_title(titulo)
        

    plt.show()
    
def get_images_path(carpetas, n_imgs):
    images_path = {}
    for carpeta in carpetas:
        for i in range(n_imgs):
            images_path[(carpeta[-7:], str(i))] = f'./dataset/{carpeta}/{carpeta[-9:]}_{str(i)}.JPEG'
    return images_path #Devuelve un diccionario en el que la clave es (carpeta, id), y los valores los paths
```

```python
n_imgs = 100
carpetas = ["autobus-n04487081", "clavos-n03804744", "coche-n02814533", "collarin-n03814639", "desatascador-n03970156",
         "gatos-n02123394", "mono-n02480495", "puentes-n04532670", "silla-n04099969", "perro-n02099601", 
            "pato-n01855672", "pizza-n07873807", "mar-n09428293", "ipod-n03584254", "platano-n07753592", "mascara_gas-n03424325", 
            "pajarita-n02883205", "mosca-n02190166", "helado-n07615774", "canon-n02950826"]
len(carpetas)
```

```python
images_path = get_images_path(carpetas, n_imgs)
len(images_path)
```

```python
# Calcular y almacenar los histogramas de texturas de todas las imágenes
histogramas_textura = [calcular_histograma_textura(imagen) for idx in images_path]

# Crear una matriz numpy para los histogramas de textura
histogramas_textura_matriz = np.array(histogramas_textura)

# Entrenar el modelo k-NN
knn = NearestNeighbors(n_neighbors=2000, algorithm='auto', metric='euclidean')
knn.fit(histogramas_textura_matriz)

new_image_path = list(images_path.values())[random.randrange(0, len(carpetas)*n_imgs - 1)]
imagen_de_consulta = cv2.cvtColor(cv2.imread(new_image_path), cv2.COLOR_BGR2RGB)
plt.imshow(imagen_de_consulta)
# Calcular el histograma de textura de la imagen de consulta
histograma_textura_consulta = calcular_histograma_textura(imagen_de_consulta).reshape(1, -1)

# Encontrar las imágenes más similares usando k-NN
distancias, indices = knn.kneighbors(histograma_textura_consulta)

# Obtener las imágenes más similares
n_similares = 8  # Puedes ajustar el número de imágenes que deseas mostrar
ids_similares = indices[0][:n_similares]
ids_similares = ids_similares.tolist()

# Mostrar las imágenes más similares utilizando las funciones definidas
display_n_similar(ids_similares, n_similares)
```
